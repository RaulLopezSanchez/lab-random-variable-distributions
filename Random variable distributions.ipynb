```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
numerical_vars = data.select_dtypes(include=[np.number])

num_columns = len(numerical_vars.columns)
num_rows = (num_columns - 1) // 3 + 1 

plt.figure(figsize=(12, 8))
for i, column in enumerate(numerical_vars.columns):
    plt.subplot(num_rows, 3, i + 1)  
    sns.histplot(numerical_vars[column], kde=True)
    plt.title(column)
    plt.tight_layout()

plt.show()

transformed_vars = np.log1p(numerical_vars) 

plt.figure(figsize=(12, 8))
for i, column in enumerate(transformed_vars.columns):
    plt.subplot(num_rows, 3, i + 1) 
    sns.histplot(transformed_vars[column], kde=True)
    plt.title(column + ' (Transformed)')
    plt.tight_layout()

plt.show()
```
